ML -SPARK 

-----------------------------


spark.mllib -> older , based on RDDs 
spark.ml -> based on dataframes 

Spark Ml pipelines -> sequence of stages to automated ml workflow by chaining together processing steps into a single repeatable pipeline .
Each step is either a transformer or an estimator .



Transformer = 1 df to another df  
tokenizer = breaks text into words


vectorassembler = multiple columns is combined into a single feature vector 

Estimator : 

stage that can fit on data to produce a transformer ( usually a model )

 e.g. LogisticRegression , RandomForestClassifier 

> PipelineModel 

Result of fitting a pipeline on data , contains the trained models and all fitted transformers 

, can be used to transform new data and make predictions 

-- Steps --

Data Ingestion and Exploration ,
Creating training and testing dataset ,
preparing features -> transformers ,
building models - estimators ,
creating pipeline ,
evaluating models ,
saving models

Now that the data is divided , we want to fill in the missing values .
There is a couple of choices we considered to fill the missing values > 1. Forward fill vs 2 . Linear interpolation .

The data will turn out as follows :
 Let's say gold closed at $1900 on Friday and opened at $1920 on Monday

Forward fill would show:
 - Friday: $1900
 - Saturday: $1900 [same price held]
 - Sunday: $1900 [same price held ]
 - Monday: $1920 [sudden jump]

Linear interpolation would show:
- Friday: $1900
- Saturday: $1906.67 [gradual increase]
- Sunday: $1913.33 [ gradual increase ]
- Monday: $1920

Financial markets hold the closing prices from Friday and show the same values on Saturday and Sunday . Thus forward fill would be most appropriate

```python
copied_train_df = train_df.copy()
# Convert 'DATE' column to datetime objects
copied_train_df['DATE'] = pd.to_datetime(copied_train_df['DATE'])

# Create a complete date range
date_rng = pd.date_range(start=copied_train_df['DATE'].min(), end=copied_train_df['DATE'].max())
complete_date_df = pd.DataFrame({'DATE': date_rng})

# Merge the complete date range with the training data
copied_train_df = pd.merge(complete_date_df, copied_train_df, on='DATE', how='left')

# Forward fill the 'VALUE' column
copied_train_df['VALUE'] = copied_train_df['VALUE'].ffill()

# Select only the 'DATE' and 'VALUE' columns
copied_train_df = copied_train_df[['DATE', 'VALUE']]

# Save the adjusted training data to a new CSV file
copied_train_df.to_csv('training_adjusted.csv', index=False)

```

Our initial setup is done , we can now start testing for stationarity which will tell what next approaches can be . It can found out by Dickey- fuller test .

```python

from statsmodels.tsa.stattools import adfuller , kpss

# Assuming 'train_df_adjusted' is your DataFrame and 'VALUE' is the column you want to test

train_df_adjusted = pd.read_csv('training_adjusted.csv')
result = adfuller(train_df_adjusted['VALUE'],autolag='AIC')

# https://stackoverflow.com/questions/47349422/how-to-interpret-adfuller-test-results
# https://www.statsmodels.org/0.6.1/generated/statsmodels.tsa.stattools.adfuller.html

kpss_result = kpss(train_df_adjusted['VALUE'])

rolling_mean = train_df_adjusted['VALUE'].rolling(window=30).mean()
rolling_std = train_df_adjusted['VALUE'].rolling(window=30).std()

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

print("\n --- Printing output --- ")

test_results = {
        'adf_test': {
            'statistic': result[0],
            'p_value': result[1],
            'No_of_Lags_Used': result[2],
            'observations_used': result[3],
            'critical_values': result[4],
            'is_stationary': result[1] < 0.05
        },
        'kpss_test': {
            'statistic': kpss_result[0],
            'p_value': kpss_result[1],
            'is_stationary': kpss_result[1] > 0.05
        },
        'rolling_statistics': {
            'mean_std': rolling_mean.std(),
            'std_std': rolling_std.std()
        }
    }
print(f"\n FINAL CONCLUSION : {test_results}")
```

---- Analysing the output results ----

A series is stationary if its statistical properties (mean, variance) remain constant over time

**ADF Test Results:**


* Statistic: 1.2012 (positive value)
* P-value: 0.9960 (very high)
* Critical values: -3.43 (1%), -2.86 (5%), -2.57 (10%)

*The high p-value (0.9960 >> 0.05) and positive test statistic strongly indicate non-stationarity*

*The test statistic (1.2012) is much higher than all critical values, which is strong evidence of a unit root*


**KPSS Test Results**:


Statistic: 9.1282 (very high)
P-value: 0.01 (very low)
Both tests agree: the series is non-stationary
The high test statistic and low p-value strongly reject the null hypothesis of stationarity


**Rolling Statistics**:


Mean standard deviation: 406.71
Standard deviation of the standard deviation: 12.53
The mean_std is much larger than std_std, indicating highly unstable mean over time

```python
import pandas as pd
import numpy as np

# Load the adjusted training data
try:
    train_df_adjusted = pd.read_csv('training_adjusted.csv')
except FileNotFoundError:
    print("Error: 'training_adjusted.csv' not found. Please make sure the file exists and the path is correct.")
    exit()

# Calculate the log difference
train_df_adjusted['Log_Difference'] = train_df_adjusted['VALUE'].diff()
train_df_adjusted['Log_Difference'] = train_df_adjusted['VALUE'].apply(lambda x: x if x==x else 0)
train_df_adjusted['Log_Difference'] =  (train_df_adjusted['VALUE'] - train_df_adjusted['VALUE'].shift()).fillna(0)
train_df_adjusted['Log_Difference'] = np.log(train_df_adjusted['VALUE']) - np.log(train_df_adjusted['VALUE']).shift()

# Display the updated DataFrame with the log difference column
print(train_df_adjusted.head(30))

# Save the updated DataFrame to a new CSV file
train_df_adjusted.to_csv('training_adjusted_with_log_diff.csv', index=False)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score
from sklearn import tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df = pd.read_csv('./data/mushrooms.csv')
df.head()
```

```python
# use LabelEncoder

labelenc = LabelEncoder()

for columns in df.columns:
    df[columns] = labelenc.fit_transform(df[columns])

df.head()
```

Classification

```python
X = df.drop(['class'],axis=1)
y= df['class']
X_train,X_test,y_train , y_test = train_test_split(X,y,test_size=0.1,random_state=33)
```

```python
# DecisionTreeClassifier

dt = DecisionTreeClassifier()
n_tree =dt.fit(X_train,y_train)
```

```python
params_choice = {
    "max_leaf_nodes" : np.linspace(10,50,30).astype(int) ,
    "max_depth" : list(range(3,9))
}


gr_search_cv = GridSearchCV(dt,params_choice,n_jobs=-1,cv=5)

gr_search_cv.fit(X_train,y_train)
print(gr_search_cv.best_estimator_)
```

```python
tuned_tree = gr_search_cv.best_estimator_
```

```python
# Bagging 

n_estimators = 100

basemodel = DecisionTreeClassifier()

m = BaggingClassifier(estimator=basemodel,n_estimators=n_estimators)

param_choice = {
    "estimator__max_depth": list(range(3,9))
}

gr_search_cv = GridSearchCV(m,param_choice,n_jobs=-1,cv = 5)
gr_search_cv.fit(X_train,y_train)

gb_m = gr_search_cv.best_estimator_

gb_m
```

```python
acc = accuracy_score(y_test,m.predict(y_test))
acc*100
```

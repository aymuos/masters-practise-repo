Learning in standardscalar 

it extracts statistical paramters or patterns from dataset 

mean and standard deviation of each parameter is computed
by using these , it can transform the training and test data to have zero mean and unit variance 



```python
import numpy as np 

from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

# Sample dataset
X = np.array([[10, 200], 
              [20, 300], 
              [30, 400], 
              [40, 500]])
scalar.fit(X)

# Print the learned mean and standard deviation
print("Mean:", scalar.mean_)  
print("Standard Deviation:", scalar.scale_)  
```

Now using fit transform 

we learn both these parameters and perform transformation 



```python
X_scaled = scalar.fit_transform(X)
print("Transformed Data:\n", X_scaled)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Sample dataset
X = np.array([[10,22], 
              [20,3], 
              [30,71], 
              [40,34]])

# StandardScaler
standard_scaler = StandardScaler()
X_standard_scaled = standard_scaler.fit_transform(X)

# MinMaxScaler (default range 0 to 1)
minmax_scaler = MinMaxScaler()
X_minmax_scaled = minmax_scaler.fit_transform(X)

# Print Results
print("Original Data:\n", X)
print("\nStandard Scaler Output:\n", X_standard_scaled)
print("\nMinMax Scaler Output:\n", X_minmax_scaled)
```
